# Creating figures

Function: matplotlib.pyplot.figure() creates a space to place the plots  
    (Parameter): figsize sets the space size  

Save subplots into variables  

Method: fig.add_subplot places each subplot into figure  
    (3 Parameters): (1,2,3)  
1. total rows of subplots we want
2. total columns of subplots we want
3. denotes which subplot variable is referencing (left-to-right, top-to-bottom)

After creating subplot, add titles to axes:  
  - set_xlabel( )
  - set_ylabel( )
 

```python
import numpy as np
import matplotlib.pyplot as plt

# Loading inflammation data csv file
data = np.loadtxt(fname='inflammation-01.csv', delimiter=',')

# Setting the size of the figure
fig = plt.figure(figsize=(10.0, 3.0))

# Creating subplots for the three descriptives
axes1 = fig.add_subplot(1,3,1)
axes2 = fig.add_subplot(1,3,2)
axes3 = fig.add_subplot(1,3,3)

# Subplot of averages
axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis=0))

# Subplot of max values
axes2.set_ylabel('max')
axes2.plot(np.max(data, axis=0))

# Subplot of min values
axes3.set_ylabel('min')
axes3.plot(np.min(data, axis=0))

# Setting the layout - giving some breathing room
fig.tight_layout()

# Saving figure as a png
plt.savefig('inflammation.png')

# Print the plots to output
plt.show()
```

## Setting Axes Ranges

    Method: .set_ylim(min, max)

## Setting the drawing style

Method: drawstyle='steps-mid'  
  
Avoid the default of interpolation to straight line between points

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt(fname='inflammation-01.csv', delimiter=',')

fig = plt.figure(figsize=(10.,3.))

ave1 = fig.add_subplot(1,3,1)
max2 = fig.add_subplot(1,3,2)
min3 = fig.add_subplot(1,3,3)

ave_data = np.mean(data, axis=0)
max_data = np.max(data, axis=0)
min_data = np.min(data, axis=0)

ave1.set_ylabel('average')
ave1.set_ylim(np.min(ave_data), np.max(ave_data)*1.1)
ave1.plot(ave_data, drawstyle='steps-mid')

max2.set_ylabel('max')
max2.set_ylim(np.min(max_data), np.max(max_data)*1.1)
max2.plot(max_data, drawstyle='steps-mid')

min3.set_ylabel('min')
min3.set_ylim(np.min(min_data), np.max(min_data)*1.1)
min3.plot(min_data, drawstyle='steps-mid')

fig.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt(fname='inflammation-01.csv', delimiter=',')

fig = plt.figure(figsize=(15., 15.))

mean_dat = np.mean(data, axis=0)
std_dat = np.std(data, axis=0)
max_dat = np.max(data, axis=0)
min_dat = np.min(data, axis=0)

ax_mean = fig.add_subplot(2,2,1)
ax_std = fig.add_subplot(2,2,2)
ax_max = fig.add_subplot(2, 2, 4)
ax_min = fig.add_subplot(2,2,3)

ax_mean.plot(mean_dat, drawstyle='steps-mid')
ax_mean.set_ylabel('average')
ax_mean.set_ylim(np.min(mean_dat), np.max(mean_dat)*1.1)

ax_std.plot(std_dat, drawstyle='steps-mid')
ax_std.set_ylabel('standard deviation')
ax_std.set_ylim(np.min(std_dat), np.max(std_dat)*1.1)

ax_max.plot(max_dat, drawstyle='steps-mid')
ax_max.set_ylabel('max')
ax_max.set_ylim(np.min(max_dat), np.max(max_dat)*1.1)

ax_min.plot(min_dat, drawstyle='steps-mid')
ax_min.set_ylabel('min')
ax_min.set_ylim(np.min(min_dat), np.max(min_dat)*1.1)

fig.tight_layout()

plt.show()
```
